```
# Get current working directory
import pandas as pd
import numpy as np
import os
import shutil

cwd = os.getcwd()
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
# Get dataset directory
dataset_dir = os.path.join(cwd, 'drive/My Drive/MaskDataset')
dataset_dir
train_dir=os.path.join(dataset_dir,'training')
os.listdir(dataset_dir)
```

```
# Import json file with labels

import json
json_labels_dir = os.path.join(dataset_dir, 'train_gt.json')

with open(json_labels_dir) as json_data:
    data = json.load(json_data)
```

```
# Create training set subfolders

values = []
for key, value in data.items():
  if value not in values:
    values.append(value)

for value in values:
  if not os.path.exists(os.path.join(train_dir, str(value))):
    os.makedirs(os.path.join(train_dir, str(value)))
```

```
# Move elements
for key, value in data.items() :
  if os.path.exists(os.path.join(train_dir, key)):
    shutil.move(os.path.join(train_dir, key),os.path.join(train_dir, str(value),key))
```
